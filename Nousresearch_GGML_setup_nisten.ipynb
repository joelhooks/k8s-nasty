```python
#█░█ █▄█ █▀█ █▀▀ █▀█ █▀█ ▄▀█ █▀█ ▄▀█ █▀▄▀█ █▀▀ ▀█▀░█▀▀ █▀█   █▀▀ █▀█ █▄░█ █▀▀ █ █▀▀ █░█ █▀█ ▄▀█ ▀█▀ █ █▀█ █▄░█░░░░░░░░░░░#|
#█▀█ ░█░ █▀▀ ██▄ █▀▄ █▀▀ █▀█ █▀▄ █▀█ █░▀░█ ██▄ ░█░░██▄ █▀▄   █▄▄ █▄█ █░▀█ █▀░ █ █▄█ █▄█ █▀▄ █▀█ ░█░ █ █▄█ █░▀█░░░░░░░░░░░#|
import os                                                                                                                #|
import torch                                                                                                             #|
#                                               ░░░                                                                      #|
# Feel free to play around with the settings here:░                                                                      #|
aimodel = "nous-hermes-llama2-13b.ggmlv3.q4_K_M.bin" # You can replace this with i.e."redmond-puffin-13b.ggmlv3.q6_K.bin"#|
prompt = "Explain how google maps api works then write a meow.ts typescript deno backend that finds the 3 closest llama farms near you."
system_prompt = "You are a hypergraph based intelligent mycelium alien lifeform that knows about everything in the galaxy."
threads = 1                           # Number of CPU threads to use, put 4 or more if you're on paid tier░░░░░░░░░░░░░░░#|
gpu_layers = 43                       # Number of model layers to load into GPU memory, leave ~15% free for context use░░#|
temperature = 0.69                    # Controls randomness: higher values (up to 1.0) make output more random░░░░░░░░░░░#|
n_predict = -1                        # Number of tokens to predict (-1 = predict indefinitely)░░░░░░░░░░░░░░░░░░░░░░░░░░#|
context_size = 3069                   # Size of the context window for the model░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░#|
color_format = "typescript"           # Output color formatting, feel free change to python, bash etc from pygmentize░░░░#|
#░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░#|
# DONT TOUCH THIS it checks of there's enough CUDA memory and sets environment variables to be used in the bash command░░#|
if torch.cuda.is_available() and torch.cuda.memory_allocated(0) > torch.cuda.get_device_properties(0).total_memory - 100*1024*1024: torch.cuda.empty_cache()
os.environ['AIMODEL'], os.environ['PROMPT'], os.environ['SYS_PROMPT'], os.environ['THREADS'], os.environ['GPU_LAYERS'], os.environ['TEMPERATURE'], os.environ['N_PREDICT'], os.environ['CONTEXT_SIZE'], os.environ['COLOR_FORMAT'] = aimodel, prompt, system_prompt, str(threads), str(gpu_layers), str(temperature), str(n_predict), str(context_size), color_format
                                                                                                                         #|
# Bash command to run the model░░░░░░░░░░░░░░░░░                                                                         #|
!stdbuf -oL ./llama.cpp/main -t $THREADS -ngl $GPU_LAYERS -m $AIMODEL --temp $TEMPERATURE -n $N_PREDICT -c $CONTEXT_SIZE \
-p "[INST] <<SYS>> $SYS_PROMPT <</SYS>> ### Instruction: $PROMPT ###Response: [/INST]" | \
stdbuf -oL fold -w 90 -s | stdbuf -oL pygmentize -s -l $COLOR_FORMAT
#░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░                                                                         #|
#█▀▀█ █░░█ ▀▀█▀▀ █▀▀█ █░░█ ▀▀█▀▀ ░ ░ ░░░░░░░░░░░░░                                                                       #|
#█░░█ █░░█ ░░█░░ █░░█ █░░█ ░░█░░ ▄ ▄ ▄░░░░░░░░░░░░                                                                       #|
#▀▀▀▀ ░▀▀▀ ░░▀░░ █▀▀▀ ░▀▀▀ ░░▀░░ █ █ █░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░░#|
```
